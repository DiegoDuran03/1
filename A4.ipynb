```python
import pandas as pd

file1 = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/adult.csv'
file2 = 'https://raw.githubusercontent.com/tec03/Datasets/main/datasets/antropometria-dataset.csv'

adult_df = pd.read_csv(file1)
antropometria_df = pd.read_csv(file2)

adult_df.head()
antropometria_df.head()
```

```python
adult_head = adult_df.head()
antropometria_head = antropometria_df.head()

adult_head, antropometria_head
```

```python
pivot_peso_est = antropometria_df.pivot_table(values='peso', index='desc_ent', aggfunc='mean')

pivot_peso_genero_est = antropometria_df.pivot_table(values='peso', index='desc_ent', columns='sexo', aggfunc='mean')

pivot_peso_est_1 = pivot_peso_est.join(pivot_peso_genero_est, lsuffix='_avg', rsuffix='_sexo')

pivot_peso_est_1
```

```python
median_peso_sex1_jal = antropometria_df[(antropometria_df['desc_ent'] == 'JALISCO') & (antropometria_df['sexo'] == 1)]['peso'].median()
median_peso_sex1_jal
```

```python
max_avg_peso_sex1 = pivot_peso_genero_est[1].max()
max_avg_peso_sex1
```

```python
adult_df['marital-status'] = adult_df['marital-status'].str.strip()
adult_df['education'] = adult_df['education'].str.strip()
```

```python
filtered_data_corrected = adult_df[(adult_df['marital-status'] == 'Divorced') & (adult_df['education'] == 'HS-grad')]
avg_age_div_hs_grad_2 = filtered_data_corrected['age'].mean()
avg_age_div_hs_grad_2
```

```python
data_male_2 = adult_df[(adult_df['sex'] == 'Male') &
                                        (adult_df['marital-status'] == 'Married-AF-spouse') &
                                        (adult_df['education'] == 'Bachelors')]
```

```python
med_hours_per_week_4 = data_male_2['hours-per-week'].median()
med_hours_per_week_4
```
