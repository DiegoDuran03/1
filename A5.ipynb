##A5: Visualizaci√≥n con Histogramas & Boxplots

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file='https://raw.githubusercontent.com/tec03/Datasets/main/datasets/antropometria-dataset.csv'

df = pd.read_csv(file)
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='sexo', y='peso', data=df)
plt.title('Boxplot del Peso por Sexo')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='sexo', y='cintura', data=df)
plt.title('Boxplot de la Cintura por Sexo')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='peso', hue='sexo', kde=True, element='step')
plt.title('Histograma del Peso por Sexo')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='cintura', hue='sexo', kde=True, element='step')
plt.title('Histograma de la Cintura por Sexo')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='cintura', y='peso', hue='sexo', data=df)
plt.title('Line Plot de Peso vs Cintura por Sexo')
plt.show()
```

```python
df['sexo'] = df['sexo'].astype('category')

plt.figure(figsize=(10, 5))
sns.scatterplot(x='cintura', y='peso', hue='sexo', data=df)
plt.title('Diagrama de Dispersi√≥n: Peso vs Cintura por Sexo')
plt.show()
```
