#Data munging / wrangling (= cleaning + transforming + reduction + enrichment)

```python
import pandas as pd
import numpy as np
```

```python
x={'Company':['Ford', 'Ford', 'VW', 'BMW', 'Cooper', 'Cooper',],
   'Stars':[1,2,np.nan,3,2,2],
   'Weight':[2,4,2,2,3,None],
   'Origin':['Mexico', 'China', 'China', 'Mexico','UK',None],
   'Length':[40,30,np.nan,56,pd.NaT,40]
   }
```

```python
type(x)
```

```python
df=pd.DataFrame(data=x)
df
```

```python
ndf=df.copy()
df
```

```python
ndf.Stars.median()
```

```python
sm=ndf.Stars.median()
sm
```

```python
ndf['Stars'].fillna(value=sm)
ndf
```

```python
ndf['Stars']=ndf['Stars'].fillna(value=sm)
ndf
```

```python
wm=ndf.Weight.mean()
wm
```

```python
ndf['Weight']=ndf['Weight'].fillna(value=wm)
ndf
```

```python
lm=ndf.Length.mean()
lm
```

```python
ndf['Length']=ndf['Length'].fillna(value=lm)
ndf
```

```python
om=ndf.Origin.mode()[1]
om
```

```python
ndf.Origin=ndf.Origin.fillna(value=om)
ndf
```
