```python
path='https://raw.githubusercontent.com/tec03/Datasets/main/datasets/reg_data.csv'
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv(path)
df.head()
```

##What is the salary of a person with age 33?

Salary=b0+b1*Age

```python
favs=['Age','Salary']
ndf=df[favs].dropna()
ndf.isnull().any()
```

```python
ndf
```

```python
import statsmodels.api as sm
from sklearn import linear_model
```

```python
x=pd.DataFrame(ndf.Age)
y=ndf.Salary
```

```python
lm=linear_model.LinearRegression()
model_lm=lm.fit(x,y)
model_lm
```

```python
model_lm.coef_
```

```python
model_lm.intercept_
```

```python
model_lm.score(x,y)
```
